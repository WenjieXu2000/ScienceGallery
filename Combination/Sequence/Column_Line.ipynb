# 柱形折线图
- 使用柱形图和折线图表示数据的范围和走势

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 随机生成两组数据
Year = np.arange(2010, 2020)  #时间轴
mean_temp = 22 # 设置10年的温度均值
temp = np.random.uniform(-3, 3, 10)  # 生成每年的温度，上下波动
```

```python
# 设定一张画布即可
fig,ax = plt.subplots(figsize=(10, 4), layout='constrained')

# 绘制每年温度的柱状图
fig1 = ax.bar(Year,temp,color='#333E50',width=0.6,label='Temperature 1')
# 在横轴上加一条水平线，横线用axhline，竖线用axvline
#ax.axhline(0,0,1,color='black',ls='--')
# 绘制左y轴方向的箭头
ax.annotate('', 
             xy=(0, 1.06), xycoords='axes fraction',
             xytext=(0, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)
# 绘制右y轴方向的箭头
ax.annotate('', 
             xy=(1.0, 1.06), xycoords='axes fraction',
             xytext=(1, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)

# 去除顶部的坐标轴
ax.spines['top'].set_visible(False)
# 修改坐标轴粗细和字体大小
ax.spines['bottom'].set_linewidth(5)
ax.tick_params(labelsize=16)


# 画出该图的姊妹图，即两个图画在一张图中
ax2 = ax.twinx()
fig2, = ax2.plot(Year,temp,'-o',color='#891C13', lw=3,label='Temperature 2') # plot的输出值需要多需要多加一个逗号解包
# 两个姊妹图的Axes需要单独设置
ax2.spines['top'].set_visible(False)
ax2.tick_params(labelsize=16)

# 利用legend中的handles参数绘制两个子图的Lable
ax2.legend(handles=[fig1, fig2], labels=['Temperature 1', 'Temperature 2'],loc='best', fontsize=15, labelcolor='Red')
plt.savefig('E:\Git Code\ScienceGallery\Picture\Column_Line.png')
```

```python
from pyecharts.charts import Sunburst
from pyecharts import options as opts
from pyecharts.globals import ThemeType
from pyecharts.render import make_snapshot

fig = Sunburst(init_opts=opts.InitOpts(width="1000px", height="600px"))
fig.add(
        "",
        data_pair=data,
        highlight_policy="ancestor",
        radius=[0, "95%"],
        sort_="null",
        levels=[
            {},
            {
                "r0": "15%",
                "r": "35%",
                "itemStyle": {"borderWidth": 2},
                "label": {"rotate": "tangential"},
            },
            {"r0": "35%", "r": "70%", "label": {"align": "right"}},
            {
                "r0": "70%",
                "r": "72%",
                "label": {"position": "outside", "padding": 3, "silent": False},
                "itemStyle": {"borderWidth": 3},
            },
        ],
    )
fig.set_global_opts(title_opts=opts.TitleOpts(title="Sunburst-"))
fig.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}"))
fig.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import PictorialBar
from pyecharts.globals import SymbolType

location = ["More than 50%", "50%~40%", "40%~30%", "Less than 30%"]
values = [17, 14, 16, 4]

fig = PictorialBar()
fig.add_xaxis(location)
fig.add_yaxis(
    "",
    values,
    label_opts=opts.LabelOpts(is_show=False),
    symbol_size=18,
    symbol_repeat="fixed",
    symbol_offset=[0, 0],
    is_symbol_clip=True,
    symbol=SymbolType.ROUND_RECT,
)
fig.reversal_axis()
fig.set_global_opts(
    xaxis_opts=opts.AxisOpts(is_show=False),
    yaxis_opts=opts.AxisOpts(
        axistick_opts=opts.AxisTickOpts(is_show=False),
        axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(
            opacity=0)),
    ),
)
fig.render_notebook()
```

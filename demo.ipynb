```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Year = np.arange(2005, 2021)
data = np.random.uniform(-26,26,16)
```

```python
fig,ax1 = plt.subplots(figsize=(10, 4), layout='constrained')

# 去除顶部的坐标轴
ax1.spines['top'].set_visible(False)

f1 = ax1.bar(Year,data,color='#333E50',width=0.6)
# 指定长度的线，横线用hlines，竖线用vlines
# 在横轴上加一条水平线，axhline
ax1.axhline(0,0,1,color='black',ls='--')
# 绘制左y轴方向的箭头
ax1.annotate('', 
             xy=(0, 1.06), xycoords='axes fraction',
             xytext=(0, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)
# 绘制右y轴方向的箭头
ax1.annotate('', 
             xy=(1.0, 1.06), xycoords='axes fraction',
             xytext=(1, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)

# 修改坐标轴粗细和字体大小
ax1.spines['bottom'].set_linewidth(5)
ax1.tick_params(labelsize=16)

# 画出该图的姊妹图，即两个图合并
ax2 = ax1.twinx()
# 两个姊妹图的Axes需要单独设置
ax2.spines['top'].set_visible(False)
ax2.tick_params(labelsize=16)

f2, = ax2.plot(Year,data,'-o',color='#891C13', lw=3)
# 绘制Lable，plot需要在输出前加一个逗号
ax2.legend(handles=[f1, f2], labels=['Annual Mean ', 'Temperatur'], loc='best', fontsize=15, labelcolor='Red')
```

```python
import matplotlib.pyplot as plt
 
variable = ['G1', 'G2', 'G3', 'G4']
data_1 = np.random.uniform(15,40,4)
data_2 = np.random.uniform(10,30,4)

# 控制柱状图的柱形粗细
width = 0.5
fig, ax = plt.subplots(figsize=(5,3), dpi=200)
ax.bar(variable, data_1, width,label='data 1')
ax.bar(variable, data_2, width, bottom=data_1,label='data 2')
ax.set_ylabel('y label')
ax.set_title('Title')
# legend的frameon可以控制标签是否显示在一个补丁中
ax.legend(ncol=2,frameon=False)
```

```python

fig,ax = plt.subplots(figsize=(8,6))
 
variable = [2016, 2017, 2018, 2019]
np.random.seed(666)
data_1 = np.random.uniform(15, 40, 4)
data_2 = np.random.uniform(5, 15, 4)
data_3 = np.random.uniform(5, 15, 4)
data_4 = np.random.uniform(10, 30, 4)
 
width = 0.5
# Hatch指粗线条绘图，而不是颜色填充
ax.bar(variable, data_1, width, label='A',color='white',hatch="//",ec='k',lw=.6)
ax.bar(variable, data_2, width,  bottom=data_2, label='B',color='gray',ec='k',lw=.6)
ax.bar(variable, data_3, width,  bottom=data_3, label='C',color='black',hatch="...",ec='k',lw=.6)
ax.bar(variable, data_4, width,  bottom=data_4, label='D',color='white',hatch="\\\\",ec='k',lw=.6)
ax.set_ylim(0,120)
ax.tick_params(direction='out',labelsize=12,length=5.5,width=1,top=False,right=False)
# legend的frameon可以控制标签是否显示在一个补丁中
ax.legend(fontsize=11,frameon=False,loc='upper center',ncol=4)
ax.set_ylabel('Numbers',fontsize=13)
ax.set_ylim(0,60)
ax.set_xlabel('Time(year)',fontsize=13)
ax.set_xlim(2015,2020)
```

```python
import numpy as np
from matplotlib import cm
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
data = pd.read_csv(r"E:\Git Code\Mathematical-modeling\Simulation practice\2018 E\Preprocess Data\Iraq.csv")
data.head()
```

```python
year = data.iloc[-16:,:]['Year']
tempmean = data.iloc[-16:,:]['平均温度/摄氏度']
all_temp_mean = np.mean(tempmean)
```

```python
x = tempmean-all_temp_mean
fig,ax1 = plt.subplots(figsize=(10, 4), layout='constrained')
# 去除上面的坐标轴
ax1.spines['top'].set_visible(False)
f1 = ax1.bar(year,x,color='#333E50',width=0.6)
# 指定长度的线，横线用hlines，竖线用vlines
# 在横轴上加一条水平线，axhline
ax1.axhline(0,0,1,color='black',ls='--')
# 绘制左y轴方向的箭头
ax1.annotate('', 
             xy=(0, 1.06), xycoords='axes fraction',
             xytext=(0, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)
# 绘制右y轴方向的箭头
ax1.annotate('', 
             xy=(1.0, 1.06), xycoords='axes fraction',
             xytext=(1, -0.06), textcoords='axes fraction',
             arrowprops=dict(facecolor='black', shrink=0.05),
             zorder=5)

# 修改坐标轴粗细和字体大小
ax1.spines['bottom'].set_linewidth(5)
ax1.tick_params(labelsize=16)

ax2 = ax1.twinx()
ax2.spines['top'].set_visible(False)
ax2.tick_params(labelsize=16)

ax2.plot(year,tempmean,'-',color='#891C13', lw=3)
f2 = ax2.scatter(year,tempmean,color='#891C13', lw=2)
ax2.legend([f1, f2], ['Annual Mean ', 'Temperatur'])

plt.savefig(r"E:\Git Code\Mathematical-modeling\Simulation practice\2018 E\Picture\趋势图.jpg",dpi=400)
```

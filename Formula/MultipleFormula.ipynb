```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fomula_1(x):
    return np.exp(x)

def fomula_2(x):
    return x*np.cos(x)

def fomula_2_twin(x):
    return x

def fomula_3(x):
    return 1/pow(x,2)

def fomula_4(x):
    return np.sin(x)
```

```python
x = np.linspace(-3.1,3.1,50)
y1 = fomula_1(x)
y2 = fomula_2(x)
y2_twin = fomula_2_twin(x)
y3 = fomula_3(x)
y4 = fomula_4(x)
```

```python
# 使用subplot2grid在特定位置创建子图
# 在第一个子图ax1中
# 第一个参数(2,3)相当于格子分成2行3列
# 第二个参数(0,0)表示该子图的开始位置
# colspan=1表示子图的列跨度为1
# rowspan=2表示子图的行跨度为2
ax1 = plt.subplot2grid((2,3),(0,0),colspan=1,rowspan=2)
# 如上设定其他子图的位置
ax2 = plt.subplot2grid((2,3),(0,1),colspan=1,rowspan=1)
ax3 = plt.subplot2grid((2,3),(0,2),colspan=1,rowspan=1)
ax4 = plt.subplot2grid((2,3),(1,1),colspan=2,rowspan=1)
```

```python
ax1 = plt.subplot2grid((2,3),(0,0),colspan=1,rowspan=2)
ax2 = plt.subplot2grid((2,3),(0,1),colspan=1,rowspan=1)
ax3 = plt.subplot2grid((2,3),(0,2),colspan=1,rowspan=1)
ax4 = plt.subplot2grid((2,3),(1,1),colspan=2,rowspan=1)
# 绘制线条
ax1.plot(x,y1,color='red')
ax1.set_yticks([0,5,10,15,20])
ax1.text(-2,20,r'$f(x)=e^{x}$',color='red')

ax2.plot(x,y2,color='green')
ax2.plot(x,y2_twin,color='dodgerblue',linestyle=':')
ax2.text(-2,2.5,r'$f(x)=xcos(x)$',color='green')
ax2.text(-2,-2.5,r'$f(x)=x$',color='dodgerblue')

# 紫色的函数可能有问题，与原图不符
ax3.plot(x,y3,color='purple')
ax3.text(-3,200,r'$f(x)=\frac{1}{x}^{2}$',color='purple')

ax4.plot(x,y4,color='goldenrod',linestyle=':')
ax4.text(-3,0.5,r'$f(x)=sin(x)$',color='goldenrod')
plt.savefig('E:\Git Code\ScienceGallery\Picture\MultiplrFormula.png')
```

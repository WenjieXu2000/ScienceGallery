```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
# 将公式定义成函数
def function(x):
    return (1/(np.sqrt(2*np.pi)))*np.exp(-pow(x,2))
```

```python
x = np.linspace(-2,2,20)  # 定义x变量的范围(-2,2), 数量20
y = function(x)
```

```python
fig,ax = plt.subplots()
# 绘制折线图, 设置线条和标记的风格,颜色,粗细
ax.plot(x,y,linestyle='--',marker='.',color='darkgreen',linewidth=2,markersize=12) 
# 重新设置x轴刻度
ax.set_xticks([-2,-1,0,1,2])
# 重新设置y轴范围
ax.set_ylim(0,0.5)
# 设置x,y轴label，Title,text
ax.set_xlabel('位置',color='blue',size=15)
ax.set_ylabel('强度',color='red',size=15)
ax.text(-0.15,0.43,'$max$',color='darkred')
# 反斜杠在 Python 字符串中是有特殊含义
# 需要加r转义
ax.set_title(r'函数$\frac{1}{\sqrt{2 \pi}} e^{-x^{2}}$',size=20)
# 设置网格
ax.grid()
plt.savefig('E:\Git Code\ScienceGallery\Picture\SimpleFormula.png')
```
